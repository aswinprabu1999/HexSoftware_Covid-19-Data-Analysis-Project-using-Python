```python
# Define a function to remove outliers from all numeric columns
def remove_outliers_all_columns(df):
    df_cleaned = df.copy()  # Create a copy of the DataFrame
    for column in df.select_dtypes(include=['int64', 'float64']).columns:  # Only process numeric columns
        Q1 = df[column].quantile(0.25)  # 25th percentile
        Q3 = df[column].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1                   # Interquartile range
        lower_bound = Q1 - 1.5 * IQR    # Lower bound
        upper_bound = Q3 + 1.5 * IQR    # Upper bound

        # Filter rows within the bounds
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    return df_cleaned

# Remove outliers from all columns
df_cleaned = remove_outliers_all_columns(df)

# Compare the original and cleaned dataset sizes
print("Original dataset size:", df.shape)
print("Cleaned dataset size:", df_cleaned.shape)
```
